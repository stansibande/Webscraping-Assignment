### <center ><font color='purple'>Step 1 Importing Libraries</font><br>

```python
import requests
import html5lib
from bs4 import BeautifulSoup
```

### <center ><font color='purple'>Step 2 Extracting the data source</font><br>

```python
#get data from scholarship page
#since the url is online, this will only work with an internet connection
url = "https://hec.ac.mw/scholarships"
response = requests.get(url)
#print get status
code=response.status_code
if code==200:
    print("Connection to data source successfull.")
else:
    print("Something went wrong, check your internet connection or URL")
```

```python
#raw page data
print(response.text)
```

### <center ><font color='purple'>Step 3 Making the data look in a readable format</font><br>

```python
#parse page so we can see the divs cleary
soup = BeautifulSoup(response.text, 'html5lib')
print(soup.prettify())
```

### <center ><font color='purple'>Step 4 Trying out extraction of Sample data using specific tags</font><br>

```python
# to find one tag
block = soup.find('h4', class_ = 'scholarship-title')
block
```

```python
block.text
```

```python
block = soup.find_all('h4', class_ = 'scholarship-title')
block
```

### <center ><font color='purple'>Step 5: Gathering All Required data</font><br>

```python
#content we are looking for is located in the below divs, trying to access to see
allData = soup.find_all('div', class_ = 'scholar-content-box')
```

```python
#create arrays to store data we will be correcting in the below loop
postingDate=[]
scholarshipName=[]
closingDate=[]
```

```python
postingDate
```

```python
scholarshipName
```

```python
closingDate
```

### <center ><font color='purple'>Step  6: Cleaning Data</font><br>

```python
#the data will be added to corresponding array, making sure unneeded text is replaced with blanks as we append
for data in allData :
    postingDate.append(data.find('p', {'class':'scholar-meta'}).text.replace('Posted on ','').replace(' by HEC Media','').strip())
    scholarshipName.append(data.find('h4', {'class':'scholarship-title'}).text.strip())
    closingDate.append(data.find_all('p')[2].text.strip().split("Closes ")[1])
```

```python
#import panda for dataframe use
import pandas as pd
```

### <center ><font color='purple'>Step 6: Showing Data</font><br>

```python
#print it out in a dataframe to display nively
df=pd.DataFrame({"Scholarship Name":scholarshipName,"Date Posted":postingDate,"Closing Date":closingDate})
df
```

### <center ><font color='purple'>Step 7: Saving Data</font><br>

```python
df.to_csv('scholarships.csv')
```
